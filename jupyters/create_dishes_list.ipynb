```python
import pandas as pd
```

```python
data = pd.read_csv('data/ingredients v1.csv')
```

```python
data.head()
```

```python
import re
from nltk.corpus import stopwords as st
from tqdm import tqdm_pandas
import tqdm
stopwords = st.words('english')

def tokenize(text):
    good_words = set()
    for word in re.split(r'[,&]', text.lower()):
        if any(map(lambda w: w not in stopwords and re.search(r'[a-z]', word), word.split())):
            good_words.add(word.strip())
    return good_words
    

foods = set()
for i in data['categories'].apply(tokenize):
    foods.update(set(i))
```

```python
import pickle
with open('data/dishes.pkl', 'wb') as file:
    pickle.dump(foods, file)
```
